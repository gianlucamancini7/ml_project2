```python
#importing necessary packages

import pandas as pd
import numpy as np
import math
from datetime import datetime, timedelta

#Opening file and saving in array
file=pd.read_csv("010217anem1.txt", header=None, error_bad_lines=False)
fileArray=np.array(file)

#Declaring all variables
dir, U, w, status, SS, ST, time=[],[],[],[],[],[],[]
u, v, theta=[],[],[]
umean, vmean, wmean=[],[],[]
ustd, vstd, wstd, thetastd=[],[],[],[]
timeplot =[]

#Initializing time step
timeStr0=fileArray[0][-1]
time0=datetime.strptime(' '.join([timeStr0]), '%d.%m.%Y %H:%M:%S')

for i,line in enumerate(fileArray):
    
    #Reading time step
    timeStr=line[-1]
    contime=datetime.strptime(' '.join([timeStr]), '%d.%m.%Y %H:%M:%S')
    time.append(contime)
       
    #calculating from raw data 
    dir_rad=math.radians(float(line[0]))
    u1=math.cos(dir_rad)*float(line[1])
    u2=math.sin(dir_rad)*float(line[1])
    
    #storing in individual array
    u.append(u1)
    v.append(u2)
    w.append(line[2])
    theta.append(line[6])
    
    #calculating stats    
    if (contime >= time0 + timedelta(seconds=3600)) or (i==fileArray.shape[0]-1):
        #mean
        umean.append(np.mean(u))
        vmean.append(np.mean(v))
        wmean.append(np.mean(w))
        #standard deviation
        ustd.append(np.std(u))
        vstd.append(np.std(v))
        wstd.append(np.std(w))
        thetastd.append(np.std(theta))
        
        #time array for plotting
        timeplot.append(contime)

        u,v,w=[],[],[]
        
        time0=contime
```
