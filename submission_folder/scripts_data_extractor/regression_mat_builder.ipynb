```python
#Imports
import pandas as pd
import numpy as np
from IPython.core import display as ICD
import glob
import os
pd.set_option('display.max_columns', 100)
```

```python
DATA_FOLDER = r'../data/average_data/'
RES_FOLDER = r'../data/'
```

List of all anemometer file is created

```python
all_anem=glob.glob(DATA_FOLDER + "anem*")
mast_anem=all_anem[:6]
top_anem=all_anem[6]
```

The column are read and the entry of `h` is added.

```python
h=np.arange(1.5,22,4)
df_list=[]
for idx,i in enumerate(mast_anem):
    df_temp=pd.read_csv(i, header=None, index_col=0, names=['direction','u','u_z','sonic_temp'])
    df_temp['h']=h[idx]
    df_list.append(df_temp)
mast_anem_df = pd.concat(df_list)
```

```python
#mast_anem_df
```

The other dataframes are opened and read. Inner joints are performed to exclude missing values.

```python
rad_df = pd.read_csv(DATA_FOLDER+'radiometer.csv',index_col=0)
temp_df = pd.read_csv(DATA_FOLDER+'surf_temp.csv',index_col=0)
top_anem_df = pd.read_csv(top_anem,index_col=0,names=['direction_top','u_top','u_top_z','sonic_temp_top'])
```

```python
#top_anem_df
```

```python
time_var_df = top_anem_df.merge(rad_df,how='inner', left_index=True,right_index=True)
time_var_df = time_var_df.merge(temp_df,how='inner', left_index=True,right_index=True)
#time_var_df.head()
```

```python
#time_var_df
```

The content of the ``Albedo`` is analyzed and discarted after being found to be noisy and not too relevant for the analysis. 

```python
tot_df=mast_anem_df.merge(time_var_df,how='inner', left_index=True,right_index=True)
tot_df = tot_df.drop(columns=['Albedo [-]'])
tot_df = tot_df.dropna(axis=0, how='any')

#tot_df
```

```python
tot_df.head()
```

```python
tot_df.to_csv(RES_FOLDER+'regression_mat_year'+'.csv', index=True, sep=',', mode='a')
```
